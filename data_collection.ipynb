**Web Scraping Example using BeautifulSoup**


```python
#BeautifulSoup

!pip install beautifulsoup4
```

*This is the site we are using "https://www.scrapethissite.com/pages/simple/"*

```python
from bs4 import BeautifulSoup

with open("scraped_data/data1.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
soup
```

```python
soup.find("h1")
soup.h3
```

```python
all_h3 = soup.find_all("h3")
all_h3
```

```python
for h3 in all_h3:
    print(h3)
```

To get contents inside the tags, say the country names, use get_text() 

```python
for h3 in all_h3:
    name = h3.get_text(strip=True)
    print(name)
    #print(h3.find_parent("div")["class"])
```

```python
for h3 in all_h3:
    name = h3.get_text(strip=True)
    print(h3.find_parent("div")["class"])
```

```python
all_h3 = soup.find_all("h3")

all_countries = []

for h3 in all_h3:
    name = h3.get_text(strip=True)
    #print(h3.find_parent("div")["class"])
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    print(population)

    print(h3.attrs)

    all_countries.append([name,population])
```

```python
all_countries
```

We can convert these details into pandas dataframe. For that : 

```python
import pandas as pd 

pd.DataFrame(all_countries)
```

Give column names for this

```python
df = pd.DataFrame(all_countries, columns=["Name", "Population"])
df 
```

Once cleaned data is found, we can convert the dataframe to csv file. For that create a folder:

```python
df.to_csv("cleaned_data/data.csv", index=False)
```
