```python
#real data

# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

#plot 
plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(non_oscar_years, non_oscar_revenue,'<--b', label ="Non Oscar Movies")
plt.plot(oscar_years, oscar_revenue, "o--r",label="Oscar Movies")
plt.title("Oscar Movies vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
