```python
#real data

# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

#plot 
plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(non_oscar_years, non_oscar_revenue,'<--b', label ="Non Oscar Movies")
plt.plot(oscar_years, oscar_revenue, "o--r",label="Oscar Movies")
plt.title("Oscar Movies vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```

**Plot Methods**

```python
movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue)
plt.title("Revenue in each Year for Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

#To display the exact revenue values above the bars
for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+5, str(oscar_revenue[i]), ha="center")  # the plus 5 is given so that the value doesn't touch the bars

#To extend the limit of y axis so that the values doesn't touch the outline 
plt.ylim(0, max(oscar_revenue)+100)  #check difference by removing this line
    
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar", color = "blue")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar", color = "orange")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
plt.tight_layout()

plt.xticks(x, years) # to show x-axis labels
```
